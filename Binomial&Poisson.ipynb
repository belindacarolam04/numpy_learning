**BINOMIAL DISTRIBUTION**

Every trial has only two outcomes

**SYNTAX**- np.random.binomial(n,p,size)


*   n-Number of trials
*   p-Probability


*   size-Number of experiments




```python
import numpy as np
import seaborn as sns
import matplotlib as plt
```

Practice-Coin Toss Simulator

```python
coin=np.random.binomial(10,0.5,50)
print(coin)
print("Maximum Number of occurences:",np.max(coin))
print("Average heads:",np.mean(coin)//1)
sns.histplot(coin)
```

**MULTONOMIAL DISTRIBUTION**

used when the trial has more than two options

**SYNTAX**-np.random.multinomial(n=10,p=[0.5,0.3,0.2],size=6)

Practice-Survey Response Simulator

```python
survey=np.random.multinomial(n=10,pvals=[0.2,0.1,0.6,0.1],size=1000)
labels=['Very Good','Good','Moderate','Poor']
print('----------------------\nFor Single Simulation\n----------------------')
print(labels[0],':',survey[0][0])
print(labels[1],':',survey[0][1])
print(labels[2],':',survey[0][2])
print(labels[3],':',survey[0][3])
print("-------------\nSurvey Report\n--------------")
total=np.sum(survey,axis=0)
print("Reviews:\n",labels[0],':',total[0])
print(labels[1],':',total[1])
print(labels[2],':',total[2])
print(labels[3],':',total[3])
print('----------\nPercentage\n----------')
print(labels[0],':',(total[0]/np.sum(total))*100)
print(labels[1],':',(total[1]/np.sum(total))*100)
print(labels[2],':',(total[2]/np.sum(total))*100)
print(labels[3],':',(total[3]/np.sum(total))*100)
sns.histplot(survey,kde=True)
```

**POISSON DISTRIBUTION**

used when counting events occuring over time or space

**SYNTAX** - np.random.poisson(lam=5,size=20)


*   lam-->lamda -average number of events




Practice-Website Traffic Analyser

*eg.Assume there are 7 users per hour. Now analyse the website traffic for 100 hours*

```python
web=np.random.poisson(7,100)
print("Outcomes\n",web)
print("------------------\nAnalysis\n--------------------")
print("Peak Visit hour:",np.argmax(web))
print("Maximum Visits at,",np.argmax(web),"th hour:",np.max(web))
print("Low Visit Hours:",np.argmin(web))
print("Minimum visits at",np.argmin(web),"th hour:",np.min(web))
sns.histplot(web)
```

**EXPONENTIAL DISTRIBUTION**

models the time until the next event happpens

**SYNTAX**- np.random.exponential(scale=2,size=10)


*   scale- average waiting time



Practice-Customer Waiting Time Analyzer

```python
waiting=np.random.exponential(scale=2,size=10)
print("Average waiting time:",np.mean(waiting).round(2))
print("Maximum Waiting Time:",np.max(waiting).round(2))
print("Minimum Waiting Time:",np.min(waiting).round(2))
print("Standard deviation:",np.std(waiting,ddof=1).round(2))
print("Variance:",np.std(waiting,ddof=1).round(2))
print("----------------\nService Levels\n----------------")
fast=np.sum(waiting<2)
moderate=np.sum((waiting>=2) & (waiting<6))
slow=np.sum(waiting>6)
print("Fast(<2min):",fast)
print("Moderate(2-6 min):",moderate)
print("Slow(>6min):",slow)
if fast>5:
  print("System is fast")
elif (slow>10):
  print("System is slow")
else:
  print("System is moderate")
print("-------------------\nNeed For Labours\n--------------")
if np.mean(waiting<2):
  print("No labours needed")
else:
  print ("Labours needed")
sns.histplot(waiting)
```
