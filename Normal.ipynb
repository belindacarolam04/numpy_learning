**NORMAL DISTRIBUTION**

  np .random.normal(loc,scale,size)


*   loc-mean
*   scale-standard deviation


*   size-number of values







IMPORTANT TERMS


*   Mean-average value
*   Standard Deviation-measure of spread from the mean(small-close to mean and vice versa)





```python
import numpy as np
```

```python
a=np.random.normal(loc=50,scale=20,size=6)
print(a)
```

```python
marks=np.random.normal(loc=70,scale=10,size=20)
print(marks)
```

PRACTICE PROBLEM-College Mark Analyser

```python
marks=np.random.normal(50,20,100)
marks = np.clip(marks, 0, 100)
print('Average Marks:',np.mean(marks))
print("Highest Mark:",np.max(marks))
print('Low Mark:',np.min(marks))
print('Students Above 80 marks:\n',marks[marks>80])
print('No of students above 80 marks:\n',np.sum(marks>80))
print('Students below 40 marks:\n',marks[marks<40])
print('No of students below 40 marks:\n',np.sum(marks<40))
print('Pass Percentage\n',(np.mean(marks>=40))*100)
```
