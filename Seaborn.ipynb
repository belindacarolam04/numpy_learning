**Seaborn Visualisation**

Uniform vs Normal Comparison

```python
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
uniform_dist=np.random.uniform(0,1,50)
sns.histplot(uniform_dist,kde=True)

#equal probability
```

```python
sns.kdeplot(uniform_dist)
```

```python
normal_dist=np.random.normal(50,20,50)
sns.histplot(normal_dist,kde=True)

#center heavy
```

```python
sns.kdeplot(normal_dist)
```
