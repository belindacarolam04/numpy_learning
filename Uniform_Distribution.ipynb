```python
import numpy as np
```

**USE OF UNIFORM**

Uniform-used for random decimal values

```python
a=np.random.uniform(0,1,size=20)
print("Random 20 values between 0 and 1\n",a)
```

```python
a=np.random.uniform(50,100,size=10)
print("Random values between 50 and 100\n",a)
```

```python
matr=np.random.uniform(size=(4,4))
print("4X4 Matrix:\n",matr)
```

PRACTICE-EMPLOYEE SALARY ANALYSER

```python
salaries=np.random.uniform(20000,100000,size=20)
print("MAXIMUM SALARY:",np.max(salaries))
print("MINIMUM SALARY:",np.min(salaries))
print("AVERAGE SALARY:",np.mean(salaries))
print("EMPLOYEES ABOVE 80000:",salaries[salaries>80000])
print("EMPLOYEES BELOW 4000:",salaries[salaries<40000])
```
