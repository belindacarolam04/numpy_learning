```python
import numpy as np
```

RANDOM AGE GENERATOR

```python
ages=np.random.randint(18,60,10)
print(ages)
```

DICEROLLS

```python
dice=np.random.randint(1,7,100)
print(dice)
```

MATRIX GENERATOR


*  np.random.rand(2,3)-positive floats
*  np.ranodm.randn(2,3)-neagative floats



```python
print(np.random.rand(2,3))
```

CHOICE


*   It is used to select from the list




```python
fruits=np.array(["apple","orange","pineapple"])
a=np.random.choice(fruits)
print(a)
```

SEED


*  same output for every run



```python
a=np.random.seed(42)
a=np.random.randint(1,7)
print(a)
```

SHUFFLE


*   shuffles the values of the list
*   return none as output if printed




```python
a=np.array([1,2,3,4,5])
np.random.shuffle(a)
print(a)
```

PERMUTATION


*   Shuffles and create copy



```python
a=np.array([1,2,3,4,5])
b=np.random.permutation(a)
print(b)
```

Practice -Lottery Number Picker
